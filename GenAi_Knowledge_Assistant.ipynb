```python


import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras

# Import GenAI library
try:
    import google.generativeai as genai
    print(" GenAI library loaded successfully!")
except ImportError:
    print(" Please install: pip install google-generativeai")
    exit()

# =============================================================================
# STEP 1:  ADDING API KEY
# =============================================================================
API_KEY = " AIzaSyCYW6eUfDg1NhymzGFK7JBl9lVhWKE6vc8"

# Configure the API
if API_KEY != " AIzaSyCYW6eUfDg1NhymzGFK7JBl9lVhWKE6vc8":
    genai.configure(api_key=API_KEY)
    print(" API Key configured!")
else:
    print(" Please paste your API key above!")

# =============================================================================
# STEP 2: LOAD THE FASHION MNIST DATASET
# =============================================================================
print("\n" + "-"*75)
print(" LOADING FASHION MNIST DATASET")
print("-"*75)
print(" DATASET TYPE: Computer Vision - Image Classification")
print(" PURPOSE: Recognize different types of clothing items")
print(" WHY IT'S GREAT FOR STUDENTS:")
print("   • Real-world application (fashion industry)")
print("   • Visual and easy to understand")
print("   • Perfect for learning image classification")
print("   • More interesting than handwritten digits!")
print("-"*75)

# Load Fashion MNIST dataset
print("\n Loading dataset...")
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()

# Class names for the 10 categories
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

print(f"\n Dataset loaded successfully!")
print(f"   Training images: {len(x_train):,}")
print(f"   Test images: {len(x_test):,}")
print(f"   Total images: {len(x_train) + len(x_test):,}")
print(f"   Image size: {x_train[0].shape[0]}x{x_train[0].shape[1]} pixels")
print(f"   Number of categories: {len(class_names)}")

# =============================================================================
# STEP 3: DATASET INFORMATION FUNCTION
# =============================================================================
def get_dataset_info():
    """Get comprehensive information about Fashion MNIST"""

    # Calculate distribution
    unique, counts = np.unique(y_train, return_counts=True)
    distribution = dict(zip(unique, counts))

    info = f"""
 FASHION MNIST DATASET - COMPLETE INFORMATION

 DATASET TYPE: Computer Vision / Image Classification
 DATA FORMAT: Grayscale Images (28x28 pixels)
 TASK: Multi-class Classification (10 categories)

 DATASET SIZE:
   • Training images: {len(x_train):,}
   • Test images: {len(x_test):,}
   • Total images: {len(x_train) + len(x_test):,}
   • Image dimensions: 28 × 28 pixels
   • Color: Grayscale (black & white)
   • Pixel values: 0 (black) to 255 (white)

 CLOTHING CATEGORIES (10 classes):
   0. T-shirt/top    - Regular t-shirts and tops
   1. Trouser        - Pants and trousers
   2. Pullover       - Sweaters and pullovers
   3. Dress          - Dresses and gowns
   4. Coat           - Coats and jackets
   5. Sandal         - Sandals and open footwear
   6. Shirt          - Formal shirts
   7. Sneaker        - Sports shoes and sneakers
   8. Bag            - Handbags and purses
   9. Ankle boot     - Boots and high footwear

 CLASS DISTRIBUTION (Training Set):
   Each category has {counts[0]:,} images (perfectly balanced!)

 MEMORY SIZE:
   • Training data: ~{x_train.nbytes / (1024*1024):.1f} MB
   • Test data: ~{x_test.nbytes / (1024*1024):.1f} MB

 WHY THIS DATASET IS PERFECT FOR STUDENTS:
   ✓ Easy to visualize and understand
   ✓ Real-world application in fashion/retail
   ✓ Challenging enough to learn from
   ✓ Not too large (trains quickly on laptops)
   ✓ Balanced classes (no bias issues)
   ✓ Great for beginner to intermediate projects
   ✓ More engaging than traditional MNIST digits

 REAL-WORLD APPLICATIONS:
   • Fashion recommendation systems
   • Automated clothing categorization for e-commerce
   • Virtual wardrobe organization apps
   • Retail inventory management
   • Fashion trend analysis
   • Smart shopping assistants
"""
    return info

# =============================================================================
# STEP 4: VISUALIZE SAMPLE IMAGES
# =============================================================================
def show_sample_images(num_samples=10):
    """Display random sample images from the dataset"""
    print(f"\n Showing {num_samples} random fashion items...\n")

    # Create grid
    cols = 5
    rows = (num_samples + cols - 1) // cols

    fig, axes = plt.subplots(rows, cols, figsize=(12, rows * 2.5))
    axes = axes.flatten() if num_samples > 1 else [axes]

    for i in range(num_samples):
        # Pick random image
        idx = np.random.randint(0, len(x_train))
        image = x_train[idx]
        label = y_train[idx]

        # Display image
        axes[i].imshow(image, cmap='gray')
        axes[i].set_title(f'{class_names[label]}',
                         fontsize=12, fontweight='bold')
        axes[i].axis('off')

    # Hide extra subplots
    for i in range(num_samples, len(axes)):
        axes[i].axis('off')

    plt.tight_layout()
    plt.show()
    print(" Images displayed!")

# =============================================================================
# STEP 5: SHOW SPECIFIC CATEGORY
# =============================================================================
def show_category_samples(category_name, num_samples=6):
    """Show multiple samples of a specific clothing category"""

    # Find category index
    try:
        category_idx = class_names.index(category_name)
    except ValueError:
        print(f" Category '{category_name}' not found!")
        print(f"   Available: {', '.join(class_names)}")
        return

    # Find all images of this category
    indices = np.where(y_train == category_idx)[0]
    selected = np.random.choice(indices, min(num_samples, len(indices)), replace=False)

    print(f"\n Showing {len(selected)} samples of: {category_name.upper()}\n")

    # Create grid
    cols = 3
    rows = (len(selected) + cols - 1) // cols

    fig, axes = plt.subplots(rows, cols, figsize=(9, rows * 3))
    axes = axes.flatten() if len(selected) > 1 else [axes]

    for i, idx in enumerate(selected):
        axes[i].imshow(x_train[idx], cmap='gray')
        axes[i].set_title(f'Sample {i+1}', fontsize=11)
        axes[i].axis('off')

    # Hide extra subplots
    for i in range(len(selected), len(axes)):
        axes[i].axis('off')

    plt.suptitle(f'{category_name.upper()} - Different Styles',
                 fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()

# =============================================================================
# STEP 6: DATASET STATISTICS
# =============================================================================
def show_statistics():
    """Display detailed dataset statistics"""
    print("\n" + "-"*75)
    print(" DETAILED DATASET STATISTICS")
    print("="*75)

    print("\n Image Properties:")
    print(f"   Image dimensions: {x_train[0].shape[0]} × {x_train[0].shape[1]} pixels")
    print(f"   Color channels: 1 (grayscale)")
    print(f"   Pixel value range: [{x_train.min()}, {x_train.max()}]")
    print(f"   Average pixel value: {x_train.mean():.2f}")
    print(f"   Pixel std deviation: {x_train.std():.2f}")

    print("\n Dataset Split:")
    print(f"   Training set: {len(x_train):,} images ({len(x_train)/(len(x_train)+len(x_test))*100:.1f}%)")
    print(f"   Test set: {len(x_test):,} images ({len(x_test)/(len(x_train)+len(x_test))*100:.1f}%)")

    print("\n Category Distribution (Training Set):")
    unique, counts = np.unique(y_train, return_counts=True)
    for idx, count in zip(unique, counts):
        percentage = (count / len(y_train)) * 100
        bar = "█" * int(percentage)
        print(f"   {class_names[idx]:12} : {count:>5,} ({percentage:>5.2f}%) {bar}")

    print("\n Dataset Balance: PERFECTLY BALANCED")
    print("   All categories have exactly 6,000 training images")

    print("\n Memory Usage:")
    train_mb = x_train.nbytes / (1024 * 1024)
    test_mb = x_test.nbytes / (1024 * 1024)
    print(f"   Training data: {train_mb:.2f} MB")
    print(f"   Test data: {test_mb:.2f} MB")
    print(f"   Total: {train_mb + test_mb:.2f} MB")

    print("-"*75)

# =============================================================================
# STEP 7: ANALYZE SPECIFIC IMAGE
# =============================================================================
def analyze_image(image_index=None):
    """Analyze a specific image in detail"""

    if image_index is None:
        image_index = np.random.randint(0, len(x_train))

    image = x_train[image_index]
    label = y_train[image_index]

    print("\n" + "-"*75)
    print(f" DETAILED IMAGE ANALYSIS - Image #{image_index}")
    print("="*75)

    print(f"\n Image Properties:")
    print(f"   Category: {class_names[label]}")
    print(f"   Label value: {label}")
    print(f"   Dimensions: {image.shape[0]}×{image.shape[1]} pixels")
    print(f"   Total pixels: {image.shape[0] * image.shape[1]:,}")
    print(f"   Pixel range: [{image.min()}, {image.max()}]")
    print(f"   Average brightness: {image.mean():.2f}")
    print(f"   Std deviation: {image.std():.2f}")

    # Display the image
    plt.figure(figsize=(6, 6))
    plt.imshow(image, cmap='gray')
    plt.title(f'{class_names[label]} (Label: {label})',
             fontsize=14, fontweight='bold')
    plt.axis('off')
    plt.show()

    print("\n Image displayed above!")
    print("="*75)

# =============================================================================
# STEP 8: ASK AI QUESTIONS
# =============================================================================
def ask_ai_question(question):
    """Ask AI assistant about the dataset"""

    if API_KEY == " AIzaSyCYW6eUfDg1NhymzGFK7JBl9lVhWKE6vc8":
        print("\n API Key not configured!")
        print("   Please paste your API key ")

        return

    print(f"\n Your Question: {question}")
    print(" AI is thinking...\n")

    # Create detailed prompt
    prompt = f"""
You are an expert AI teaching assistant specializing in Computer Vision and
Machine Learning, helping students learn about the Fashion MNIST dataset.

Dataset Context:
{get_dataset_info()}

Student's Question: {question}

Please provide a clear, educational answer that:
1. Is easy for students to understand
2. Uses simple language and examples
3. Relates to real-world applications when possible
4. Encourages further learning
5. Is accurate and technically correct

Answer in a friendly, encouraging tone suitable for students.
"""

    try:
        model = genai.GenerativeModel('gemini-pro')
        response = model.generate_content(prompt)

        print("-"*75)
        print(" AI TEACHER'S ANSWER:")
        print("-"*75)
        print(response.text)
        print("-"*75)

    except Exception as e:
        print(f"\n Error: {str(e)}")
        print("\n Troubleshooting:")
        print("   1. Check your API key is correct")
        print("   2. Verify internet connection")
        print("   3. Make sure you have API quota left")
        print("   4. Visit: https://makersuite.google.com")

# =============================================================================
# STEP 9: GET LEARNING GUIDE
# =============================================================================
def get_learning_guide():
    """Get AI-generated learning guide for students"""

    if API_KEY == " AIzaSyCYW6eUfDg1NhymzGFK7JBl9lVhWKE6vc8":
        print("\n Please add your API key first!")
        return

    print("\n Generating Student Learning Guide...")


    prompt = f"""
As an AI teaching assistant, create a comprehensive learning guide for
students working with the Fashion MNIST dataset.

Dataset Information:
{get_dataset_info()}

Please provide:
1. What is Fashion MNIST and why it's great for learning
2. Step-by-step learning path for students
3. Simple project ideas they can build
4. Recommended model architectures (CNN basics)
5. Tips for improving model accuracy
6. Common mistakes to avoid
7. Resources for further learning

Make it encouraging, structured, and perfect for beginners!
"""

    try:
        model = genai.GenerativeModel('gemini-pro')
        response = model.generate_content(prompt)

        print("-"*75)
        print(" STUDENT LEARNING GUIDE")
        print("-"*75)
        print(response.text)
        print("-"*75)

    except Exception as e:
        print(f" Error: {str(e)}")

# =============================================================================
# STEP 10: PROJECT IDEAS GENERATOR
# =============================================================================
def get_project_ideas():
    """Get AI-generated project ideas for students"""

    if API_KEY == " AIzaSyCYW6eUfDg1NhymzGFK7JBl9lVhWKE6vc8":
        print("\n Please add your API key first!")
        return

    print("\n Generating Project Ideas...\n")

    prompt = f"""
Generate 5 creative and educational project ideas for students using the
Fashion MNIST dataset. Make them progressively more challenging.

Dataset: Fashion MNIST - 10 categories of clothing items

For each project, provide:
- Project title
- Difficulty level (Beginner/Intermediate/Advanced)
- Description
- Learning outcomes
- Suggested tools/libraries

Make ideas practical, fun, and educational!
"""

    try:
        model = genai.GenerativeModel('gemini-pro')
        response = model.generate_content(prompt)

        print("-"*75)
        print(" STUDENT PROJECT IDEAS")
        print("-"*75)
        print(response.text)
        print("-"*75)

    except Exception as e:
        print(f"Error: {str(e)}")

# =============================================================================
# STEP 11: MAIN DEMO
# =============================================================================
def run_demo():
    """Run complete demonstration"""

    print("\n" + "-"*75)
    print(" GENAI KNOWLEDGE ASSISTANT - DEMO MODE")
    print("-"*75)

    # Show dataset info
    print(get_dataset_info())

    # Show statistics
    show_statistics()

    # Show sample images
    show_sample_images(10)

    # Show specific category
    show_category_samples('Sneaker', 6)

    # Analyze specific image
    analyze_image()

    # AI Features
    if API_KEY != " AIzaSyCYW6eUfDg1NhymzGFK7JBl9lVhWKE6vc8":
        print("\n" + "-"*75)
        print("AI-POWERED LEARNING FEATURES")
        print("-"*75)

        # Example questions
        ask_ai_question("Why is Fashion MNIST better than regular MNIST for students?")
        ask_ai_question("What's the simplest neural network I can build for this dataset?")
        ask_ai_question("How do I preprocess these images before training?")

        # Learning guide
        get_learning_guide()

        # Project ideas
        get_project_ideas()
    else:
        print("\n" + "-"*75)
        print(" AI FEATURES DISABLED")
        print("-"*75)
        print("Add your API key to unlock:")
        print("  • Ask questions about the dataset")
        print("  • Get personalized learning guidance")
        print("  • Receive project ideas")
        print("  • Get AI tutoring on concepts")

        print("-"*75)

    print("\n Demo completed!")

# =============================================================================
# STEP 12: INTERACTIVE MODE
# =============================================================================
def interactive_mode():
    """Interactive Q&A mode for students"""

    print("\n" + "-"*75)
    print(" INTERACTIVE LEARNING MODE")
    print("-"*75)
    print("Ask me anything about Fashion MNIST and image classification!")
    print("\nCommands:")
    print("  • Type your question to ask AI")
    print("  • 'stats' - Show dataset statistics")
    print("  • 'samples' - Show random images")
    print("  • 'category <name>' - Show specific clothing type")
    print("  • 'analyze' - Analyze a random image")
    print("  • 'guide' - Get learning guide")
    print("  • 'projects' - Get project ideas")
    print("  • 'categories' - List all categories")
    print("  • 'exit' - Quit")
    print("-"*75)

    while True:
        user_input = input("\n Student: ").strip()

        if not user_input:
            continue
        elif user_input.lower() == 'exit':
            print("\n Keep learning! Goodbye!")
            break
        elif user_input.lower() == 'stats':
            show_statistics()
        elif user_input.lower() == 'samples':
            num = input("   How many? (1-20): ").strip()
            try:
                show_sample_images(min(int(num), 20))
            except:
                show_sample_images(10)
        elif user_input.lower().startswith('category '):
            category = user_input[9:].strip()
            show_category_samples(category)
        elif user_input.lower() == 'analyze':
            analyze_image()
        elif user_input.lower() == 'guide':
            get_learning_guide()
        elif user_input.lower() == 'projects':
            get_project_ideas()
        elif user_input.lower() == 'categories':
            print("\n👕 Available Categories:")
            for i, name in enumerate(class_names):
                print(f"   {i}. {name}")
        else:
            ask_ai_question(user_input)

# =============================================================================
# MAIN EXECUTION
# =============================================================================
if __name__ == "__main__":
     # Run demo
    run_demo()

    # Ask about interactive mode
    print("\n" + "-"*75)
    choice = input("Want to try Interactive Learning Mode? (yes/no): ").strip().lower()

    if choice in ['yes', 'y']:
        interactive_mode()
    else:
        print("\n You can call these functions anytime:")
        print("   • show_statistics()")
        print("   • show_sample_images(10)")
        print("   • show_category_samples('Dress', 6)")
        print("   • analyze_image(100)")
        print("   • ask_ai_question('your question')")
        print("   • get_learning_guide()")
        print("   • get_project_ideas()")
        print("   • interactive_mode()")
```
